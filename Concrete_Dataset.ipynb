```python
#using OLS
import pandas as pd
filename='Concrete_Data.xls'
df=pd.read_excel(filename)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
X_train, X_test, y_train, y_test=train_test_split(X, y,random_state=0)
linreg=LinearRegression().fit(X_train,y_train)
print("Training Accuracy:",linreg.score(X_train,y_train))
print("Testing Accuracy:",linreg.score(X_test,y_test))
```

```python
#Ridge Regression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
X_train, X_test, y_train, y_test=train_test_split(X, y,random_state=0)

linreg=Ridge(alpha=0.01).fit(X_train,y_train)
print("Training Accuracy:",linreg.score(X_train,y_train))
print("Testing Accuracy:",linreg.score(X_test,y_test))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso
X = df.iloc[:, :-1]  # All columns except the last one
y = df.iloc[:, -1]   # Only the last column

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=66)

LR = Lasso(alpha = 0.05).fit(X_train, y_train)
print("Training Accuracy: ",LR.score(X_train, y_train))
print("Testing Accuracy: ", LR.score(X_test, y_test))
```
