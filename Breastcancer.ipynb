```python
#breastcancer dataset
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn .tree import DecisionTreeClassifier
breastcancer = load_breast_cancer()
X_train,X_test,y_train,y_test=train_test_split(breastcancer['data'],breastcancer['target'] ,random_state=42)
#trainmodel
#preprunning
model=DecisionTreeClassifier(max_depth=3).fit(X_train,y_train)
print("training accuracy",model.score(X_train,y_train))
print("testing accuracy",model.score(X_train,y_train))
from sklearn import tree
#helps you visualize
tree.plot_tree(model)
import matplotlib.pyplot as plt
plt.show()
#preprunning is best advised but comes withb its limitations good at level 4
print(model.feature_importances_)
```
