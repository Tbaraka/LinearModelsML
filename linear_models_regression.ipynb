```python
#Linear models -OLS
#retrieving the datasets
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
my_data=load_diabetes()
X_train,X_test,y_train,y_test=train_test_split(my_data.data,my_data.target,random_state=0)
#train the data
liReg=LinearRegression().fit(X_train,y_train)#specifies which algorithm to use and trains
print("Training accuracy" ,liReg.score(X_train,y_train) )
print("test accuracy" ,liReg.score(X_test,y_test) )
#the model is overfitting
```

```python
#RidgeRegression
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
my_data=load_diabetes()
X_train,X_test,y_train,y_test=train_test_split(my_data.data,my_data.target,random_state=0)
#train the data
liReg=Ridge().fit(X_train,y_train)#specifies which algorithm to use and trains
print("Training accuracy" ,liReg.score(X_train,y_train) )
print("test accuracy" ,liReg.score(X_test,y_test))
```

```python
#RidgeRegression uisng alpha to minimize the overfitting by regularizing the impact of the gradient
#uses pythogras theorem
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
my_data=load_diabetes()
X_train,X_test,y_train,y_test=train_test_split(my_data.data,my_data.target,random_state=0)
#train the data
liReg=Ridge(alpha=0.01).fit(X_train,y_train)#default value for alpha is 0.01
print("Training accuracy" ,liReg.score(X_train,y_train) )
print("test accuracy" ,liReg.score(X_test,y_test))
```

```python
#lasso regression
#manhattan
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso
my_data=load_diabetes()
X_train,X_test,y_train,y_test=train_test_split(my_data.data,my_data.target,random_state=0)
#train the data
liReg=Lasso(alpha=0.2).fit(X_train,y_train)#default value for alpha is 0.01
print("Training accuracy" ,liReg.score(X_train,y_train) )
print("test accuracy" ,liReg.score(X_test,y_test))
print("number of features",sum(liReg.coef_!=0))#write the coefficinets that are not 0
```
